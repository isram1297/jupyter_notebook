### Ejemplos de Expresiones Aritméticas

1. Suma: 2 + 3 = 5
2. Resta: 5 - 2 = 3
3. Multiplicación: 4 * 6 = 24
4. División: 10 / 2 = 5
5. Exponente: 2^3 = 8
6. Raíz Cuadrada: √9 = 3
7. Módulo (Residuo de la división): 10 % 3 = 1
8. Paréntesis para agrupar operaciones: (4 + 6) * 2 = 20
9. Orden de operaciones (PEMDAS/BODMAS):
   - Parentesis
   - Exponentes
   - Multiplicación y División (de izquierda a derecha)
   - Suma y Resta (de izquierda a derecha)


```python
# Multiplicación de números
num1 = 5
num2 = 3
resultado_multiplicacion = num1 * num2
print("El resultado de la multiplicación es:", resultado_multiplicacion)

# Suma de números
num3 = 10
num4 = 7
resultado_suma = num3 + num4
print("El resultado de la suma es:", resultado_suma)
```
